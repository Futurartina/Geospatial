```python
import pandas as pd
import folium
from folium import Icon

sf_df = pd.read_csv("../data/result_sf_df.csv")
csv_file_path = "../data/result_sf_df.csv"
sf_df
# Reads the df related to San Francisco.
```

```python
san_francisco_lat = 37.773972
san_francisco_lon = -122.431297
# Defines coordenates of San Francisco.
```

```python
san_francisco = folium.Map(location=[san_francisco_lat, san_francisco_lon], zoom_start=12)  
san_francisco
# Prints an initial map of San Francisco. 
```

```python
def get_marker_icon(Type):

    if Type == 'Starbucks':
        return Icon (
                icon = "fa-solid fa-mug-saucer",
                prefix = "fa",
                color = "beige",
                icon_color = "white")

    elif Type == 'Vegan rest':
         return Icon (
                icon = "fa-solid fa-carrot",
                prefix = "fa",
                color = "white",
                icon_color = "orange")
       
    elif Type == 'Preschool':
         return Icon (
                icon = "fa-solid fa-school",
                prefix = "fa",
                color = "green",
                icon_color = "white")
        
    else:
           return Icon (
                icon = "fa-solid fa-plane",
                prefix = "fa",
                color = "blue",
                icon_color = "white")
# Defines the format of each different marker in the map.        
```

```python
for index, row in sf_df.iterrows():
    marker_icon = get_marker_icon(row['Type'])
    folium.Marker(
        location=[row['lat'], row['lon']],
        popup=row['name'],
        icon=marker_icon
    ).add_to(san_francisco)
# Checks the Type in the df and adds each marker to the df. 
```

```python
san_francisco
# Prints the new map of San Francisco. 
```

```python
outer_radius = 4000
inner_radius = 2000
# Defines an outer_radius and an inner_radius for the circle. 
```

```python
folium.Circle(
    location=[san_francisco_lat, san_francisco_lon],
    radius=outer_radius,
    color='gray',  # Color of the circle outline
    fill=True,
    fill_color='lightgray',  # Fill color for the circle
    fill_opacity=0.2  # Opacity of the circle fill
).add_to(san_francisco)

folium.Circle(
    location=[san_francisco_lat, san_francisco_lon],
    radius=inner_radius,
    color='cadetblue',  # Color of the circle outline
    fill=True,
    fill_color='lightblue',  # Fill color for the circle
    fill_opacity=0.4  # Opacity of the circle fill
).add_to(san_francisco)

san_francisco

# Adds an outer and an inner circle based on the indicated radius. 
```

```python
san_francisco.save("san_francisco.html")
```
